```python
import numpy as np

K=9200
Premium_Call=100

Interval=1000
ST=np.arange(K-Interval,K+Interval)

ST
```

```python
Payoff_LongCall=np.maximum(ST-K,0)-Premium_Call
Payoff_ShortCall=-Payoff_LongCall
```

```python
import matplotlib.pyplot as plt

plt.plot(ST,Payoff_LongCall)
plt.plot(ST,Payoff_ShortCall)

plt.show()
```

```python
import numpy as np 

K1=9200
K2=9500

Premium_Call=100
Premium_Put=80

Interval=1000
ST=np.arange(K1-Interval,K1+Interval)



ST
```

```python
Payoff_LongTopStrangle=-(np.maximum(K1-ST,0))-Premium_Put-(np.maximum(ST-K2,0))+Premium_Call
```
